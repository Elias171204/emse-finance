## Let's first build the volatilty skew for a given maturity date :

The bs_forward_price function is the function that takes:
* The forward price F
* The strike K
* the maturity date T
* sigma :
* df : df = $\exp(-rT)$
* option type : if it's a call or a put

it returns the payoff of the option prices with Black scholes

```python
def bs_forward_price(F, K, T, sigma, df, option_type):
    if T <= 0 or sigma <= 0:
        return np.nan

    d1 = (np.log(F / K) + 0.5 * sigma**2 * T) / (sigma * np.sqrt(T))
    d2 = d1 - (sigma * np.sqrt(T))

    if option_type == "call":
        return df * (F * norm.cdf(d1) - K * norm.cdf(d2))
    else:
        return df * (K * norm.cdf(-d2) - F * norm.cdf(-d1))
```

The implied_vol_forward function is the function that takes : 

* the price mid
* F the price forward
* K the strike
* T the maturity date
* df = $\exp(-rT)$
* option type if it is a call or a put

This functions returns the implied volatility given by the BS formula. In other words it finds the implied volatility given the price of the options given by the market and using the black scholes formula.

Given the fact that :

* $d_1 = \frac{\ln\left(\frac{F}{K}\right) + \tfrac{1}{2}\sigma^2 T}{\sigma \sqrt{T}}$
* $d_2 = d_1 - \sigma \sqrt{T}$

We are seeking the $\sigma$ that verifies : $Price_{BS}(\sigma) - Price_{mkt} = 0$

In fact we use brentq in order to get the right $\sigma$ which is the implied volatility

```python
def implied_vol_forward(price, F, K, T, df, option_type):
    if price <= 0 or T <= 0:
        return np.nan

    if option_type == "call":
        lower = max(df * (F - K), 0.0)
        upper = df * F
    else:
        lower = max(df * (K - F), 0.0)
        upper = df * K

    if not (lower < price < upper):
        return np.nan

    def objective(sigma):
        return bs_forward_price(F, K, T, sigma, df, option_type) - price

    try:
        return brentq(objective, 1e-6, 5.0)
    except ValueError:
        return np.nan
```

The function volatility_skew takes the implied volatility of all the data that we have.

```python
def volatility_skew(exp_date, data, atm_band=0.01):
    df_exp = data[data["exp_date"] == exp_date].copy()


    df_exp["k"] = np.log(df_exp["strike"] / df_exp["F"])

    df_exp["iv_market"] = df_exp.apply(
        lambda row: implied_vol_forward(
            price=row["mid"],
            F=row["F"],
            K=row["strike"],
            T=row["T"],
            df=row["df"],
            option_type=row["type.1"],
        ),
        axis=1
    )

    df_exp = df_exp.dropna(subset=["iv_market"])

    left = df_exp[(df_exp["k"] < -atm_band) & (df_exp["type.1"] == "put")]
    right = df_exp[(df_exp["k"] > atm_band) & (df_exp["type.1"] == "call")]

    atm = df_exp[(df_exp["k"].abs() <= atm_band)].copy()
    if ("bid" in atm.columns) and ("ask" in atm.columns):
        atm["spread"] = atm["ask"] - atm["bid"]
        atm = atm.sort_values(["spread", "k"]).drop_duplicates(subset=["strike"], keep="first")
        atm = atm.drop(columns=["spread"])
    else:

        atm = atm.sort_values("k").drop_duplicates(subset=["strike"], keep="first")

    out = (
        pd.concat([left, atm, right], axis=0, ignore_index=True)
        .sort_values("k")
        .reset_index(drop=True)
    )

    return out[["k", "strike", "type.1", "iv_market"]]
```

```python
def build_surface_points(data, min_T=7/365, atm_band=0.01):
    exp_dates = sorted(data["exp_date"].unique())

    pieces = []
    for exp in exp_dates:
        skew = volatility_skew(exp, data, atm_band=atm_band).copy()

        T_val = data.loc[data["exp_date"] == exp, "T"].iloc[0]
        if T_val < min_T:
            continue

        skew["T"] = T_val
        pieces.append(skew[["T", "k", "iv_market"]])

    surface_pts = pd.concat(pieces, ignore_index=True)

    surface_pts = surface_pts[(surface_pts["iv_market"] > 0) & (surface_pts["iv_market"] < 3.0)]
    return surface_pts
```

```python
def interpolate_iv_surface(surface_pts, k_min=-0.5, k_max=0.25, nk=80, nT=40, method="linear"):

    T_grid = np.linspace(surface_pts["T"].min(), surface_pts["T"].max(), nT)
    k_grid = np.linspace(k_min, k_max, nk)
    TT, KK = np.meshgrid(T_grid, k_grid, indexing="xy")


    points = np.column_stack([surface_pts["T"].values, surface_pts["k"].values])
    values = surface_pts["iv_market"].values
    IV = griddata(points, values, (TT, KK), method=method)


    if np.isnan(IV).any():
        IV_near = griddata(points, values, (TT, KK), method="nearest")
        IV = np.where(np.isnan(IV), IV_near, IV)

    return T_grid, k_grid, IV
```

```python
def plot_iv_surface_3d(T_grid, k_grid, IV):
    TT, KK = np.meshgrid(T_grid, k_grid, indexing="xy")

    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection="3d")
    ax.plot_surface(TT, KK, IV, rstride=1, cstride=1, linewidth=0)
    ax.set_xlabel("T (years)")
    ax.set_ylabel("k = ln(K/F)")
    ax.set_zlabel("IV")
    ax.set_title("SPY market implied volatility surface")
    plt.show()
```

```python

surface_pts = build_surface_points(data, min_T=7/365, atm_band=0.01)
T_grid, k_grid, IV = interpolate_iv_surface(surface_pts, k_min=-0.5, k_max=0.25, nk=90, nT=50, method="linear")

TT, KK = np.meshgrid(T_grid, k_grid, indexing="xy")


fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111, projection="3d")

surf = ax.plot_surface(TT, KK, IV, linewidth=0, antialiased=True)

ax.set_xlabel("Maturity T (years)")
ax.set_ylabel("log-moneyness k = ln(K/F)")
ax.set_zlabel("Market implied volatility")
ax.set_title(f"SPY implied volatility surface (market)")

plt.show()

```

# The Heston Model in the case of vanilla options (EU call) :

The Heston model is made to rebuild the implied volatility surface.

The Heston Model corresponds to choosing from:

$$
dS_t = \mu_t S_t \, dt + \sqrt{v_t}\, S_t \, dZ_1
$$

$$
dv_t = \alpha(S_t, v_t, t)\, dt + \eta\, \beta(S_t, v_t, t)\, \sqrt{v_t}\, dZ_2
$$

such that $\alpha = - \lambda (v_t - \bar{v})$ and $ \beta = 1$, where $ \lambda$ is the speed of reversion of $v_t$ to its long-term mean $\bar{v}$

with:
$
\langle dZ_1 \, dZ_2 \rangle = \rho \, dt
$
where:

* $Z_1, Z_2 \text{: are brownian motions}$
* $\rho \in [-1,1] \quad \text{: correlation between random stock price returns an changes in v (leverage effect)}$
* $\eta \quad \text{: volatility of volatility}$


Remark : Here $v_t$ is the instantaneaous variance of the log return under the riskless measure. The integrale from 0 to T of this variable is the realized variance.

In the literature, here are the names of the variales that we are seeking to fit the model:


Under the **risk-neutral measure** $ \mathbb{Q} $, the Heston model is defined by the following system of stochastic differential equations:

$$
\begin{aligned}
dS_t &= (r - q)\, S_t \, dt + \sqrt{v_t}\, S_t \, dW_t^{(S)}, \\
dv_t &= \kappa (\theta - v_t)\, dt + \sigma \sqrt{v_t}\, dW_t^{(v)}, \\
d\langle W^{(S)}, W^{(v)} \rangle_t &= \rho \, dt.
\end{aligned}
$$



Our goal now is to fit the Heston model and so to find the right values for:
$\kappa$;
$\theta$;
$\rho$;
$\sigma$;
$v_0$.


```python
def heston_cf_lns(u, T, S0, r, q, kappa, theta, sigma, rho, v0):
    """
    Characteristic function of ln(S_T) under the risk-neutral Heston model:
        dS_t = (r-q) S_t dt + sqrt(v_t) S_t dW1_t
        dv_t = kappa (theta - v_t) dt + sigma sqrt(v_t) dW2_t
        corr(dW1, dW2) = rho

    Returns: phi(u) = E[ exp(i u ln(S_T)) ]
    """
    i = 1j
    u = np.asarray(u, dtype=np.complex128)

    x0 = np.log(S0)
    a = kappa * theta
    b = kappa

    # Standard "u^2 + i u" term for ln(S_T) CF
    d = np.sqrt((rho * sigma * i * u - b)**2 + (sigma**2) * (u**2 + i*u))

    # Stable choice of g (common stable convention)
    g = (b - rho * sigma * i * u - d) / (b - rho * sigma * i * u + d)

    exp_dt = np.exp(-d * T)

    # Use log1p to reduce cancellation when g*exp(-dT) ~ 1 or g ~ 1
    log_term = np.log1p(-g * exp_dt) - np.log1p(-g)

    C = i*u*(x0 + (r - q)*T) + (a / sigma**2) * ((b - rho*sigma*i*u - d)*T - 2.0*log_term)
    D = ((b - rho*sigma*i*u - d) / sigma**2) * ((1.0 - exp_dt) / (1.0 - g * exp_dt))

    return np.exp(C + D * v0)
```

```python
def heston_call_price(S0, K, T, r, q, kappa, theta, sigma, rho, v0):
    i = 1j
    lnK = np.log(K)

    phi_mi = heston_characteristic_function(-i, T, S0, r, q, kappa, theta, sigma, rho, v0)

    def integrand(u, j):
        u = np.asarray(u, dtype=np.complex128)

        if j == 1:
            phi = heston_characteristic_function(u - i, T, S0, r, q, kappa, theta, sigma, rho, v0) / phi_mi
        else:  # j == 2
            phi = heston_characteristic_function(u, T, S0, r, q, kappa, theta, sigma, rho, v0)

        return np.real(np.exp(-i * u * lnK) * phi / (i * u))

    eps = 1e-8
    UMAX = 100.0

    P1 = 0.5 + (1 / np.pi) * quad(lambda u: integrand(u, 1), eps, UMAX, limit=200)[0]
    P2 = 0.5 + (1 / np.pi) * quad(lambda u: integrand(u, 2), eps, UMAX, limit=200)[0]

    return S0 * np.exp(-q * T) * P1 - K * np.exp(-r * T) * P2
```

```python
S0 = 100.0
K  = 110.0
T  = 1.0
r  = 0.03
q  = 0.01

kappa = 2.0
theta = 0.04
sigma = 0.50
rho   = -0.70
v0    = 0.04

price = heston_call_price(S0, K, T, r, q, kappa, theta, sigma, rho, v0)
print(f"Call price (Heston): {price:.6f}")
```
