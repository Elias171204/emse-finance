## Chapter 0: Quick Remind:

Let's quickly remind all the maths needed for the rest of the notebook. Here is the Black & Scholes formula:

$$
\frac{\partial V}{\partial t} + \frac{1}{2}\,\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} + r S \frac{\partial V}{\partial S} - r V
= 0
$$

with the following assumptions:

* V is the derivative price
* S : stock price follows a geometric browninan motion where $ \sigma $ and $ \mu $ are constant
* No restriction on short selling
* No transaction fees. Every underlying asset is perfectly divisable
* No arbitrage opportunity
* Continuous market
* r : risk free rate is constant and independant of the maturity of the derivative


## Chapter 1 : Stochastic Volatility and Local Volatility:

From chapter 0, we clearly understand that constant volatility assumption is too restrictive. It then makes sens to model volatility as a random variable as we will show later on with the GLE example.

Stochastic volatility (SV) models are useful because they explain in a self-consistent way why options with different strikes and expirations have different Black-Scholes implied volatilities—that is, the "volatility smile".

In fact to price derivatives we don't use the historical volatility, because we want to price the future not the past. Then we have this definition of implied volatitlity: Implied volatility is the constant volatility parameter that makes a pricing model reproduce observed market option prices.

This is why we need to introduce models in order to predict the implied volatility. 


### IBKR data function:

```python
def ibkr_download(
    ib,
    ticker: str,
    start: str,
    end: str,
    interval: str,
    exchange: str,
    currency: str,
    primaryExchange: str,
    chunk: str,
    useRTH: bool = False,
    sleep_s: float = 0.6
) -> pd.DataFrame:
    """
    Simple wrapper style yfinance, mais IBKR.
    - ib : instance IB déjà connectée
    - start/end : 'YYYY-MM-DD'
    - interval : ex '15 mins'
    """

    def to_naive(ts: pd.Timestamp) -> pd.Timestamp:
        ts = pd.to_datetime(ts)
        if getattr(ts, "tzinfo", None) is not None:
            ts = ts.tz_localize(None)
        return ts

    start_dt = to_naive(start)
    end_dt   = to_naive(end)

    contract = Stock(ticker, exchange, currency, primaryExchange=primaryExchange)
    ib.qualifyContracts(contract)

    dfs = []
    current_end = end_dt

    while current_end > start_dt:
        bars = ib.reqHistoricalData(
            contract,
            endDateTime=current_end.strftime("%Y%m%d %H:%M:%S"),
            durationStr=chunk,            # "1 Y" (si pacing: "6 M")
            barSizeSetting=interval,      # "15 mins"
            whatToShow="TRADES",
            useRTH=useRTH,
            formatDate=1
        )

        if not bars:
            break

        df = util.df(bars).rename(columns={"date": "Datetime"})
        df["Datetime"] = pd.to_datetime(df["Datetime"])

        # IMPORTANT: rendre tz-naive si IB renvoie une timezone
        if hasattr(df["Datetime"].dt, "tz") and df["Datetime"].dt.tz is not None:
            df["Datetime"] = df["Datetime"].dt.tz_localize(None)

        dfs.append(df)

        current_end = df["Datetime"].min()

        ib.sleep(sleep_s)

    if not dfs:
        return pd.DataFrame()

    out = (pd.concat(dfs, ignore_index=True)
             .drop_duplicates(subset=["Datetime"])
             .sort_values("Datetime")
             .reset_index(drop=True))

    return out[(out["Datetime"] >= start_dt) & (out["Datetime"] <= end_dt)]
```

Commentaire:



Let's now focus on the Wilmott model which is the following, where S is the stock price and v its variance, then we have:

$$
dS_t = \mu_t S_t \, dt + \sqrt{v_t}\, S_t \, dZ_1
$$

$$
dv_t = \alpha(S_t, v_t, t)\, dt + \eta\, \beta(S_t, v_t, t)\, \sqrt{v_t}\, dZ_2
$$
with

$$
\langle dZ_1 \, dZ_2 \rangle = \rho \, dt
$$

where:

$$
Z_1, Z_2 \text{: are brownian motions}
$$
$$
\rho \in [-1,1] \quad \text{: correlation between random stock price returns an changes in v (leverage effect)}
$$
$$
\eta \quad \text{: volatility of volatility}
$$


In the black Scholes case, there is only one source of randomness which is the stock price, which can be hedged with stock. In our case, random changes in volatility also need to be hedged in order to form a riskless portfolio. Demonstration of the formula is at page 5. Here is the formula:

$$
\frac{\partial V}{\partial t} + \frac{1}{2}\, v S^2 \frac{\partial^2 V}{\partial S^2} + \rho\, \eta\, v\, \beta\, S \frac{\partial^2 V}{\partial v\, \partial S} + \frac{1}{2}\, \eta^2 v \beta^2 \frac{\partial^2 V}{\partial v^2} + r S \frac{\partial V}{\partial S} - r V
= -\left(\alpha - \phi\, \beta \sqrt{v}\right)\frac{\partial V}{\partial v}
\tag{1.3}
$$

Where $\alpha$ and $\beta$ are the function mentionned earlier, alpha is the drift and beta the volatility function for instantaneous variance


### The Market Price of Volatility:

Let's define the risk neutral drift as:

$$ \alpha' = \alpha - \beta \sqrt{v} \phi$$

## Chapter 2 : The Heston Model:

The Heston model is made to rebuild the implied volatility surface.

The Heston Model corresponds to choosing from:

$$
dS_t = \mu_t S_t \, dt + \sqrt{v_t}\, S_t \, dZ_1
$$

$$
dv_t = \alpha(S_t, v_t, t)\, dt + \eta\, \beta(S_t, v_t, t)\, \sqrt{v_t}\, dZ_2
$$

such that $\alpha = - \lambda (v_t - \bar{v})$ and $ \beta = 1$, where $ \lambda$ is the speed of reversion of $v_t$ to its long-term mean $\bar{v}$

Remark : Here $v_t$ is the instantaneaous variance of the log return under the riskless measure. The integrale from 0 to T of this variable is the realized variance.
