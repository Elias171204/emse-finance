```python
# Imports needed:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

# 1. Let's build the yield curve:

```python
# Here is a first example where we will take random example for the interest rates:

d = {'Maturity' : [1,2,3,4,5,8,10,15,20,25,30],
     'Interest Rate' : [0.01,0.013, np.nan, np.nan, 0.02, np.nan, 0.024,np.nan, np.nan, 0.03, np.nan]
    }
data = pd.DataFrame(data=d)
```

```python
data
```

```python
maturity = data['Maturity']
interest_rate = data['Interest Rate']
```

```python
# Let's visualize our yield curve:

plt.scatter(maturity, interest_rate)
plt.title('Real Interest rates')
plt.xlabel('Maturity')
plt.ylabel('Interest Rate')
```
